```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import LSTM, Dense,Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing import sequence
import pickle
```

```python
df = pd.read_csv("spam.csv", encoding="latin")
```

```python
df.head()
```

```python
df = df[["v1","v2"]]
```

```python
df.head()
```

```python
df["v1"].value_counts()
```

```python
# get X as feature and Y as target 
```

```python
lab_map = {"ham":0, "spam":1} #dict
```

```python
Y = df["v1"].map(lab_map).values
```

```python
Y
```

```python
X = df["v2"].values
```

```python
X
```

```python
# convert to sequences
# pad sequence
# train
```

```python
max_words = 1000
mytokenizer = Tokenizer(num_words=max_words,lower=True, split=" ") #tokenizáló beálítása
```

```python
mytokenizer.fit_on_texts(X) # tokenek létrehozása az X-ből
```

```python
len(mytokenizer.word_counts)
```

```python
mytokenizer.word_counts
```

```python
text_tokenized = mytokenizer.texts_to_sequences(X) # szövegből szekvenciák létrehozása
```

```python
text_tokenized
```

```python
max_len = 50
sequences = sequence.pad_sequences(text_tokenized,maxlen=max_len)
```

```python
sequences
```

```python
# Build model
```

```python
model = Sequential()
model.add(Embedding(max_words, 20, input_length=max_len))
model.add(LSTM(64))
model.add(Dense(1, activation="sigmoid"))
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(sequences,Y,batch_size=128,epochs=10,
          validation_split=0.2)
```

```python
# Predict on new test data
```

```python
inp_test_seq = "WINNER! U win a 500 prize reward & free entry to FA cup final tickets! Text FA to 34212 to receive award"
```

```python
test_sequences = mytokenizer.texts_to_sequences(np.array([inp_test_seq]))
test_sequences_matrix = sequence.pad_sequences(test_sequences,maxlen=max_len)
```

```python
model.predict(test_sequences_matrix)
```

```python
import os
MODEL_PATH = os.path.sep.join(['/content/', "SPAM_detector.h5"])
print("[INFO] saving object detector model...")
model.save(MODEL_PATH, save_format="h5")
```

```python
from tensorflow.keras.models import load_model
print("[INFO] loading SPAM detector...")
model2 = load_model(MODEL_PATH)
```

```python
model2.predict(test_sequences_matrix)
```
